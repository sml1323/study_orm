## 1. SQLAlchemy와 ORM의 이해

첫 번째 주제는 **SQLAlchemy와 ORM이 무엇인지** 이해하는 거예요.

음, 데이터베이스가 거대한 도서관이라고 상상해 보세요. 책(데이터)들이 가지런히 정리되어 있죠. 이 도서관에서 책을 찾고, 새로 넣고, 수정하고, 버리는 일(데이터 조작)을 해야 한다면 어떻게 할까요?

가장 전통적인 방법은 사서에게 직접 "책장 25번 칸에 있는 'SQLAlchemy 완전 정복' 책 좀 가져다주세요!"라고 말하는 거예요. 이게 바로 **SQL**을 직접 사용해서 데이터베이스에 명령하는 것과 같아요. 즉, SQL은 데이터베이스와 대화하기 위한 '언어'라고 생각하면 됩니다.

그런데 우리가 파이썬 코드를 작성하고 있는데, 데이터베이스와 대화할 때마다 갑자기 SQL 언어로 바꿔서 이야기해야 한다면 좀 불편하겠죠? 이때 **SQLAlchemy**와 **ORM(객체 관계형 매핑)**이 등장합니다!

### SQLAlchemy는 무엇인가요?

**SQLAlchemy**는 파이썬으로 데이터베이스를 다룰 수 있게 해주는 라이브러리예요. 마치 우리가 복잡한 도서관 규칙을 몰라도 쉽게 책을 다룰 수 있게 도와주는 '도서관 관리 시스템'과 같아요. SQLAlchemy는 크게 두 가지 방식으로 데이터베이스와 소통할 수 있게 해줘요.

1.  **SQL Expression Language**: SQL 문을 파이썬 코드로 유연하게 조합해서 만들 수 있게 해줘요. SQL을 직접 쓰는 것과 비슷하지만 파이썬의 장점을 활용할 수 있죠.
2.  **ORM(Object Relational Mapper)**: 바로 우리가 집중해서 배울 부분이에요!

### ORM(객체 관계형 매핑)이란?

**ORM**은 Object Relational Mapping의 약자로, 우리가 파이썬에서 만든 **객체(Object)**를 데이터베이스의 **테이블(Relational)**과 연결(Mapping)시켜주는 기술이에요.

다시 도서관 비유로 돌아가 볼까요? ORM은 우리가 '책'이라는 개념을 파이썬 코드 안에서 `Book`이라는 클래스로 만들면, 이 `Book` 객체가 자동으로 도서관의 '책' 테이블에 있는 데이터와 연결되게 해주는 마법 같은 도구예요.

* `my_book = Book(title='SQLAlchemy 정복', author='김튜터')` 이렇게 파이썬 객체를 만들면,
* ORM이 알아서 이 정보를 데이터베이스의 '책' 테이블에 새로운 행으로 추가해 주는 거죠!

**ORM의 장점**은 다음과 같아요.

* **생산성 향상**: SQL 문을 직접 작성하지 않고 파이썬 코드로 데이터베이스를 다룰 수 있어 개발 속도가 빨라져요.
* **유지보수 용이**: 파이썬 객체 지향적으로 코드를 작성하기 때문에 코드가 더 깔끔하고 이해하기 쉬워져요.
* **데이터베이스 독립성**: 데이터베이스 종류(MySQL, PostgreSQL, SQLite 등)가 바뀌어도 코드 수정이 최소화되거나 필요 없을 수 있어요.

자, 이제 SQLAlchemy와 ORM이 무엇인지 대략적인 감이 오시나요? 혹시 궁금한 점이 있으시거나, 더 자세히 알아보고 싶은 부분이 있으신가요?

## 2\. 모델(Model) 정의하기

두 번째 주제는 **모델을 정의하는 방법**이에요. 파이썬 클래스를 사용해서 데이터베이스 테이블을 만들고, 그 안에 데이터를 넣고, 조회하고, 수정하고, 삭제하는 기본적인 작업을 배워볼 거예요.

### 테이블과 컬럼 매핑: 파이썬 클래스로 데이터베이스 만들기

ORM의 핵심은 파이썬 클래스 하나가 데이터베이스의 **테이블** 하나와 대응되고, 클래스 안의 변수들이 테이블의 \*\*컬럼(Column)\*\*에 대응된다는 거예요.

마치 우리가 집을 짓기 위해 설계도를 그리는 것과 같아요. 설계도(파이썬 클래스)에 방 개수, 창문 위치 등을 정해놓으면, 실제 집(데이터베이스 테이블)이 그 설계도대로 지어지는 거죠.

SQLAlchemy에서 모델을 정의하는 기본적인 방법은 다음과 같아요:

1.  **`declarative_base()`**: 모든 모델 클래스의 "기반"이 되는 클래스를 만들어요. 이 클래스를 상속받아서 우리의 모델 클래스를 만들 거예요.
2.  **`Column`**: 테이블의 각 컬럼을 정의할 때 사용해요. 컬럼의 타입(정수, 문자열 등)과 제약 조건(기본 키, 널 불허 등)을 지정할 수 있어요.
3.  **`Integer`, `String`, `Boolean` 등**: SQLAlchemy가 제공하는 다양한 데이터 타입이에요.

간단한 `User` 모델을 만들어 볼까요? 이 모델은 `id`, `name`, `email` 세 가지 컬럼을 가질 거예요.

```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import sessionmaker, declarative_base

# 1. 데이터베이스 연결 설정
# SQLite 메모리 데이터베이스를 사용해서 별도 파일 없이 바로 테스트할 수 있어요.
DATABASE_URL = "sqlite:///:memory:"
engine = create_engine(DATABASE_URL)

# 2. 모든 모델 클래스의 기반이 되는 Base 생성
Base = declarative_base()

# 3. User 모델 정의: 테이블과 컬럼 매핑
class User(Base):
    __tablename__ = 'users' # 데이터베이스에서 사용할 테이블 이름

    id = Column(Integer, primary_key=True) # 기본 키 (Primary Key)
    name = Column(String(50), nullable=False) # 문자열, null 허용 안 함
    email = Column(String(100), unique=True, nullable=False) # 문자열, 고유 값, null 허용 안 함

    def __repr__(self):
        return f"<User(id={self.id}, name='{self.name}', email='{self.email}')>"

# 4. 데이터베이스에 테이블 생성
Base.metadata.create_all(engine)

print("User 테이블이 성공적으로 생성되었습니다!")
```

위 코드를 실행하면 `users`라는 이름의 테이블이 메모리 데이터베이스에 생성됩니다. `id`는 정수형 기본 키, `name`은 최대 50자의 문자열이고 비워둘 수 없으며, `email`은 최대 100자의 문자열이면서 고유해야 하고 비워둘 수 없는 컬럼이 되는 거죠.

## 2\. 모델(Model) 정의하기 (계속)

아주 좋습니다\! 이제 모델을 어떻게 정의하는지 배웠으니, 이 모델을 사용해서 실제 데이터를 다루는 방법을 알아볼게요. 바로 **데이터 추가(Create), 조회(Read), 수정(Update), 삭제(Delete)**, 줄여서 **CRUD** 작업입니다.

이 과정은 마치 도서관에서 책을 빌리고, 읽고, 내용을 고치고, 반납하는 것과 같아요. ORM 덕분에 이 모든 작업을 파이썬 객체를 다루듯이 할 수 있죠.

### 세션(Session)이란?

CRUD 작업을 하기 전에 중요한 개념이 하나 있어요. 바로 \*\*세션(Session)\*\*입니다. 세션은 데이터베이스와의 모든 상호작용이 일어나는 임시 공간이자 통로라고 생각하면 돼요.

  * 우리가 도서관에서 책을 빌리거나 반납할 때, 일단 사서에게 가서 "이 책 빌려주세요" 또는 "이 책 반납할게요"라고 말하죠? 사서가 그 작업을 기록하고 최종적으로 처리해줍니다.
  * 여기서 **세션**은 바로 이 **사서**와 같은 역할을 해요. 우리가 파이썬 객체로 데이터 작업을 하면, 그 작업들은 바로 데이터베이스에 반영되는 것이 아니라 일단 세션에 "예약"됩니다.
  * 그리고 `session.commit()`을 호출해야 세션에 예약된 모든 변경사항이 한꺼번에 데이터베이스에 **영구적으로 저장**돼요.
  * 만약 작업 도중에 문제가 생겨서 모든 것을 취소하고 싶다면 `session.rollback()`을 사용해서 이전 상태로 되돌릴 수 있어요.

### 데이터 추가(Create)

새로운 데이터를 데이터베이스에 추가하는 것은 파이썬 객체를 만들고 세션에 추가한 다음 커밋하는 과정이에요.



````python
from sqlalchemy import create_engine, Column, Integer, String, Text
from sqlalchemy.orm import sessionmaker, declarative_base

# 데이터베이스 연결 설정
DATABASE_URL = "sqlite:///:memory:"
engine = create_engine(DATABASE_URL)

# Base 생성
Base = declarative_base()

class Product(Base):
    __tablename__ = "products"
    id = Column(Integer, primary_key=True)
    name = Column(String(100), nullable=False)
    price = Column(Integer, nullable=False)
    description = Column(Text, nullable=True)

    def __repr__(self):
        return f"<Product(id={self.id}, name='{self.name}', price={self.price})>"

# 테이블 생성
Base.metadata.create_all(engine)

# 세션 생성
Session = sessionmaker(bind=engine)
session = Session()

print("--- 데이터 추가(CREATE) ---")

# 1. Product 객체 생성
new_product = Product(name='스마트폰', price=800000, description='최신형 스마트폰')
print(f"새로운 상품 객체: {new_product}")

# 2. 세션에 객체 추가 (아직 데이터베이스에 저장된 건 아님)
session.add(new_product)
print(f"세션에 추가 후 (아직 커밋 전): {new_product}") # id가 아직 할당 안 됐을 수도 있어요.

# 3. 변경사항 커밋 (데이터베이스에 영구 저장)
session.commit()
print(f"커밋 후 (id 할당됨): {new_product}") # 이제 id가 할당된 것을 볼 수 있어요.

# 여러 개의 객체 한 번에 추가
product2 = Product(name='무선 이어폰', price=150000, description='고음질 무선 이어폰')
product3 = Product(name='스마트 워치', price=300000, description='건강 관리 스마트 워치')
session.add_all([product2, product3]) # add_all을 사용하면 여러 객체를 한 번에 추가할 수 있어요.
session.commit()
print(f"여러 개 추가 후: {product2}, {product3}")

session.close() # 세션 사용이 끝나면 닫아주는 것이 좋아요.
print("\n데이터 추가 완료!")


# `Product` 객체를 만들고 `session.add()`로 세션에 추가한 다음 `session.commit()`으로 데이터베이스에 저장되는 과정을 볼 수 있을 거예요. 특히 `id`가 커밋 후에 자동으로 할당되는 것을 확인해 보세요\!
````



### **활동: 직접 상품 데이터 추가하기**

그럼 이제 `Product` 테이블이 잘 생성되었다고 가정하고, 이전에 배운 내용을 활용해서 직접 상품 데이터를 추가해 볼까요?

다음 코드를 참고해서, 여러분이 생각하는 새로운 상품 2\~3개를 만들어서 데이터베이스에 추가해 보세요.

```python
from sqlalchemy import create_engine, Column, Integer, String, Text
from sqlalchemy.orm import sessionmaker, declarative_base

# 데이터베이스 연결 설정
DATABASE_URL = "sqlite:///:memory:"
engine = create_engine(DATABASE_URL)

# Base 생성
Base = declarative_base()

class Product(Base):
    __tablename__ = "products"
    id = Column(Integer, primary_key=True)
    name = Column(String(100), nullable=False)
    price = Column(Integer, nullable=False)
    description = Column(Text, nullable=True)

    def __repr__(self):
        return f"<Product(id={self.id}, name='{self.name}', price={self.price})>"

# 1. 테이블 생성 (이 부분이 실제 데이터베이스에 테이블을 만듭니다!)
Base.metadata.create_all(engine)

# 2. 세션 생성
Session = sessionmaker(bind=engine)
session = Session()
```

## 2\. 모델(Model) 정의하기 (계속)

아주 좋아요\! 이제 데이터베이스에 새로운 상품들을 추가하는 방법을 마스터하셨네요. 이제는 데이터베이스에 잘 저장된 정보들을 다시 \*\*조회(Read)\*\*하는 방법을 배워볼 차례입니다. 데이터베이스에서 필요한 책을 찾아오는 과정이라고 생각하면 돼요.

SQLAlchemy에서 데이터를 조회하는 것은 크게 두 가지 방식으로 이루어져요.

1.  **`session.get()`**: \*\*기본 키(Primary Key)\*\*를 이용해서 특정 하나의 객체를 빠르게 찾아올 때 사용해요. 마치 도서관에서 책 번호(고유한 ID)를 알려주고 그 책을 바로 가져오는 것과 같죠.
2.  **`session.query()` 또는 `session.scalars()` (SQLAlchemy 2.0 스타일)**: 더 복잡한 조건으로 여러 객체를 조회할 때 사용해요. 특정 제목의 책을 찾거나, 특정 저자의 책을 모두 찾는 등 다양한 조건을 적용할 수 있죠. (여기서는 주로 `session.query()`를 사용하겠지만, 최신 방식인 `session.scalars()`도 간단히 언급할게요.)

### 데이터 조회(Read)

먼저 `session.get()`을 사용해서 `id`로 특정 상품을 조회해볼게요.



````python

from sqlalchemy import create_engine, Column, Integer, String, Text
from sqlalchemy.orm import sessionmaker, declarative_base

# 데이터베이스 연결 설정
DATABASE_URL = "sqlite:///:memory:"
engine = create_engine(DATABASE_URL)

# Base 생성
Base = declarative_base()

class Product(Base):
    __tablename__ = "products"
    id = Column(Integer, primary_key=True)
    name = Column(String(100), nullable=False)
    price = Column(Integer, nullable=False)
    description = Column(Text, nullable=True)

    def __repr__(self):
        return f"<Product(id={self.id}, name='{self.name}', price={self.price})>"

# 테이블 생성
Base.metadata.create_all(engine)

# 세션 생성 및 데이터 추가 (조회를 위해 데이터가 필요해요!)
Session = sessionmaker(bind=engine)
session = Session()

try:
    # 조회할 데이터 미리 추가
    product1 = Product(name='마우스', price=35000, description='사무용 마우스')
    product2 = Product(name='스피커', price=15000, description='사무용 스피커')
    product3 = Product(name='웹캠', price=70000, description='화상 회의용 웹캠')
    session.add_all([product1, product2, product3])
    session.commit()
    print("초기 상품 데이터 추가 완료!")

    print("\n--- 데이터 조회(READ) ---")

    # 1. Primary Key로 단일 객체 조회 (session.get())
    # id가 2인 상품을 찾아볼까요?
    retrieved_product = session.get(Product, 2)
    if retrieved_product:
        print(f"ID가 2인 상품: {retrieved_product}")
    else:
        print("ID 2인 상품을 찾을 수 없습니다.")

    # 2. 모든 객체 조회 (session.query().all())
    # from sqlalchemy import select (SQLAlchemy 2.0 권장 방식)
    # products = session.scalars(select(Product)).all()
    all_products = session.query(Product).all() # 모든 Product 객체를 리스트로 반환
    print("\n모든 상품:")
    for product in all_products:
        print(f"- {product}")

    # 3. 조건에 맞는 객체 조회 (session.query().filter_by())
    # 가격이 30000원 이상인 상품을 찾아볼까요?
    expensive_products = session.query(Product).filter(Product.price >= 30000).all() # filter_by도 가능
    # filter_by 사용 예시: session.query(Product).filter_by(name='마우스').first()
    print("\n가격이 30000원 이상인 상품:")
    for product in expensive_products:
        print(f"- {product}")

    # 4. 특정 조건의 첫 번째 객체 조회 (.first())
    # 이름이 '스피커'인 상품 중 첫 번째 것을 찾아볼까요?
    speaker = session.query(Product).filter_by(name='스피커').first()
    if speaker:
        print(f"\n이름이 '스피커'인 상품: {speaker}")
    else:
        print("\n'스피커'를 찾을 수 없습니다.")

except Exception as e:
    session.rollback()
    print(f"조회 중 오류 발생: {e}")
finally:
    session.close()
    print("\n데이터 조회 완료!")
````

위 코드를 실행해 보면, 다양한 방식으로 데이터를 조회하는 방법을 익힐 수 있을 거예요. 특히 `session.get()`, `session.query().all()`, `session.query().filter()`, `session.query().filter_by()`, `.first()`의 차이점을 눈여겨보세요.

## 3\. `relationship`으로 테이블 관계 설정하기

데이터베이스는 단순히 독립적인 테이블들의 모음이 아니에요. 실제 세상의 정보들처럼, 테이블들도 서로 **관계**를 맺고 있답니다. 예를 들어, 우리 주변에는 이런 관계들이 많죠.

  * **한 명의 작가**는 **여러 권의 책**을 쓸 수 있어요. (작가 ↔ 책: 일대다 관계)
  * **여러 명의 학생**은 **하나의 수업**을 들을 수 있어요. (학생 ↔ 수업: 다대일 관계)
  * **한 권의 책**은 **여러 명의 저자**에 의해 쓰일 수 있고, **한 명의 저자**는 **여러 권의 책**을 쓸 수 있어요. (책 ↔ 저자: 다대다 관계)

이런 관계들을 데이터베이스에 표현하기 위해 우리는 \*\*외래 키(Foreign Key)\*\*라는 것을 사용해요.

### 외래 키(Foreign Key) 개념

**외래 키**는 한 테이블의 특정 컬럼이 다른 테이블의 \*\*기본 키(Primary Key)\*\*를 참조하는 것을 말해요. 쉽게 말해, 서로 다른 테이블을 연결해 주는 "끈" 또는 "다리" 역할을 합니다.

예를 들어, '작가' 테이블과 '책' 테이블이 있다고 해봅시다.

  * **작가 테이블 (Authors)**: `id` (PK), `name`
  * **책 테이블 (Books)**: `id` (PK), `name`, `author_id` (FK)

여기서 `Books` 테이블의 `author_id` 컬럼이 `Authors` 테이블의 `id` 컬럼을 참조하는 **외래 키**가 되는 거예요. `author_id` 덕분에 우리는 특정 책이 어떤 작가의 책인지 알 수 있고, 특정 작가가 어떤 책들을 썼는지도 찾아낼 수 있죠.

외래 키를 설정할 때는 `ForeignKey`를 사용해요.
`Column(Integer, ForeignKey('authors.id'))` 이런 식으로요.

### `relationship`의 기본 사용법 (일대다, 다대일)

SQLAlchemy의 `relationship`은 이 외래 키로 연결된 두 모델 객체 간의 관계를 파이썬 코드 상에서 자연스럽게 탐색할 수 있게 해주는 마법 같은 기능이에요. 마치 복잡한 지도를 보지 않고도 "이 작가의 모든 책을 보여줘\!" 또는 "이 책의 작가가 누구야?"라고 말하면 바로 알려주는 비서와 같죠.

가장 흔한 관계인 \*\*일대다(One-to-Many)\*\*와 **다대일(Many-to-One)** 관계를 예시로 들어볼게요.

**시나리오**: 쇼핑몰에서 상품(Product)은 반드시 하나의 카테고리(Category)에 속한다고 가정해봅시다.

  * **Category**: `id`, `name` (예: 전자제품, 의류, 식품)
  * **Product**: `id`, `name`, `price`, `category_id` (어떤 카테고리에 속하는지)

이 경우, **하나의 카테고리**는 **여러 개의 상품**을 가질 수 있고 (일대다), **하나의 상품**은 **하나의 카테고리**에만 속할 수 있죠 (다대일).

이제 코드로 어떻게 `Category`와 `Product` 모델을 정의하고 `relationship`을 설정하는지 살펴볼게요.



````python

from sqlalchemy import create_engine, Column, Integer, String, Text, ForeignKey
from sqlalchemy.orm import sessionmaker, declarative_base, relationship
from sqlalchemy.orm import Mapped, mapped_column # 2.0 스타일도 함께 보여드릴게요! (선택적)
from typing import List, Optional # 2.0 스타일 타입 힌팅을 위한 임포트

# 데이터베이스 연결 설정
DATABASE_URL = "sqlite:///:memory:"
engine = create_engine(DATABASE_URL)

# Base 생성 (1.x 스타일)
Base = declarative_base()

# 혹은 2.0 스타일 Base (이 경우 Mapped, mapped_column 사용 가능)
# class Base(DeclarativeBase):
#     pass

# 1. Category 모델 정의 (일 "One" 부분)
class Category(Base):
    __tablename__ = 'categories'
    id = Column(Integer, primary_key=True)
    name = Column(String(50), nullable=False, unique=True)

    # 이 부분이 중요해요! '일' 쪽에서 '다' 쪽을 참조하는 relationship
    # 'products'는 이 카테고리에 속한 모든 Product 객체들의 리스트가 될 거예요.
    # back_populates='category'는 Product 모델의 'category' 속성과 연결된다는 의미예요.
    products = relationship('Product', back_populates='category')
    # 2.0 스타일: products: Mapped[List["Product"]] = relationship(back_populates="category")

    def __repr__(self):
        return f"<Category(id={self.id}, name='{self.name}')>"

# 2. Product 모델 정의 (다 "Many" 부분)
class Product(Base):
    __tablename__ = 'products'
    id = Column(Integer, primary_key=True)
    name = Column(String(100), nullable=False)
    price = Column(Integer, nullable=False)
    description = Column(Text, nullable=True)

    # 외래 키 설정: 'categories' 테이블의 'id'를 참조합니다.
    category_id = Column(Integer, ForeignKey('categories.id'), nullable=False)
    # 2.0 스타일: category_id: Mapped[int] = mapped_column(ForeignKey("categories.id"))

    # 이 부분이 중요해요! '다' 쪽에서 '일' 쪽을 참조하는 relationship
    # 'category'는 이 Product 객체가 속한 Category 객체가 될 거예요.
    # back_populates='products'는 Category 모델의 'products' 속성과 연결된다는 의미예요.
    category = relationship('Category', back_populates='products')
    # 2.0 스타일: category: Mapped["Category"] = relationship(back_populates="products")

    def __repr__(self):
        return f"<Product(id={self.id}, name='{self.name}', price={self.price}, category_id={self.category_id})>"

# 테이블 생성
Base.metadata.create_all(engine)

# 세션 생성
Session = sessionmaker(bind=engine)
session = Session()

try:
    print("--- relationship 기본 사용법 ---")

    # 카테고리 생성
    electronics = Category(name='전자제품')
    clothing = Category(name='의류')
    session.add_all([electronics, clothing])
    session.commit()
    print(f"카테고리 생성: {electronics}, {clothing}")

    # 상품 생성 및 카테고리에 연결
    # 방법 1: 외래 키 직접 지정
    phone = Product(name='스마트폰', price=800000, category_id=electronics.id)
    tshirt = Product(name='반팔 티셔츠', price=25000, category_id=clothing.id)
    session.add_all([phone, tshirt])
    session.commit()
    print(f"상품 생성 (FK 직접 지정): {phone}, {tshirt}")

    # 방법 2: relationship 속성 사용 (훨씬 직관적이고 추천하는 방법!)
    # ORM이 알아서 category_id를 채워줍니다.
    laptop = Product(name='노트북', price=1500000, category=electronics)
    jeans = Product(name='청바지', price=70000, category=clothing)
    session.add_all([laptop, jeans])
    session.commit()
    print(f"상품 생성 (relationship 사용): {laptop}, {jeans}")

    print("\n--- relationship을 이용한 객체 탐색 ---")

    # 1. 카테고리에서 소속 상품 조회 (일대다)
    print(f"\n'{electronics.name}' 카테고리의 상품들:")
    # electronics.products는 자동으로 이 카테고리에 연결된 Product 객체들의 리스트를 반환
    for p in electronics.products:
        print(f"- {p.name} (가격: {p.price})")

    # 2. 상품에서 소속 카테고리 조회 (다대일)
    print(f"\n'{phone.name}'의 카테고리:")
    # phone.category는 자동으로 이 상품에 연결된 Category 객체를 반환
    print(f"- {phone.category.name}")
    print(f"'{jeans.name}'의 카테고리:")
    print(f"- {jeans.category.name}")

except Exception as e:
    session.rollback()
    print(f"관계 설정 및 탐색 중 오류 발생: {e}")
finally:
    session.close()

print("\nrelationship 기본 사용법 및 외래 키 개념 학습 완료!")
````



위 코드를 찬찬히 살펴보세요.

  * `ForeignKey('categories.id')`: `Product` 모델에 `category_id` 컬럼이 `categories` 테이블의 `id`를 참조하도록 외래 키를 정의했습니다.
  * **`relationship()`**:
      * `Category` 모델에서는 `products = relationship('Product', back_populates='category')` 를 통해 **`Category` 객체에서 이 카테고리에 속하는 `Product` 객체들을 리스트로 바로 접근**할 수 있게 했어요.
      * `Product` 모델에서는 `category = relationship('Category', back_populates='products')` 를 통해 **`Product` 객체에서 자신이 속한 `Category` 객체를 바로 접근**할 수 있게 했고요.
      * `back_populates`는 양방향 관계를 설정할 때 사용하며, 서로의 관계 속성 이름을 알려주는 역할을 해요. 이게 있어야 한쪽에서 변경하면 다른 쪽에서도 자동으로 반영됩니다.

## 3\. `relationship`으로 테이블 관계 설정하기 (계속)

````python

from sqlalchemy import create_engine, Column, Integer, String, Text, ForeignKey
from sqlalchemy.orm import sessionmaker, declarative_base, relationship

# 데이터베이스 연결 설정
DATABASE_URL = "sqlite:///:memory:"
engine = create_engine(DATABASE_URL)

# Base 생성
Base = declarative_base()

# ----------------------------------------
# 1. 여기에 Author 모델을 정의하세요.
# class Author(Base):
#     __tablename__ = 'authors'
#     # ... 컬럼 및 relationship 정의
#     # ... __repr__ 정의

# 2. 여기에 Book 모델을 정의하세요.
# class Book(Base):
#     __tablename__ = 'books'
#     # ... 컬럼 및 외래 키, relationship 정의
#     # ... __repr__ 정의
# ----------------------------------------

class Author(Base):
    __tablename__ = 'authors'
    id = Column(Integer, primary_key=True)
    name = Column(String(100), nullable=False)
    
    books = relationship('Book', back_populates='author')

    def __repr__(self):
        # print 대신 return을 사용하고, 문자열을 직접 반환합니다.
        return f"<Author(id={self.id}, name='{self.name}')>"

class Book(Base):
    __tablename__ = 'books'
    id = Column(Integer, primary_key=True)
    title = Column(String(200), nullable=False)
    # published_year는 비워둘 수 있다고 했으니 nullable=True로 변경하는 게 더 좋습니다.
    published_year = Column(Integer, nullable=True) 
    
    author_id = Column(Integer, ForeignKey('authors.id'), nullable=False)

    author = relationship('Author', back_populates='books')
    
    def __repr__(self):
        # print 대신 return을 사용하고, 문자열을 직접 반환합니다.
        # 책 정보에 'books' 대신 'title'을 사용하는 것이 더 적합합니다.
        return f"<Book(id={self.id}, title='{self.title}', published_year={self.published_year}, author_id={self.author_id})>"

# 테이블 생성
Base.metadata.create_all(engine)

# 세션 생성
Session = sessionmaker(bind=engine)
session = Session()

try:
    print("--- Author-Book 관계 설정 및 데이터 다루기 활동 시작! ---")

    # ----------------------------------------
    # 3. 여기에 작가와 책 데이터를 추가하세요.
    # author1 = Author(...)
    # book1 = Book(..., author=author1)
    # ...
    # session.add_all([...])
    # session.commit()
    # ----------------------------------------

    author1 = Author(name="김철수")
    author2 = Author(name="이영희")
    session.add_all([author1, author2])
    session.commit()

    print(f"생성된 작가: {author1}, {author2}")
    # 각 작가에게 2권 이상의 책을 연결하여 생성하고 커밋
    # 이때 author_id 대신 relationship 속성인 'author'에 Author 객체 자체를 할당합니다.
    book1 = Book(title="파이썬 기초", published_year=2023, author=author1) # author_id는 ORM이 자동으로 채움
    book2 = Book(title="고급 데이터베이스", published_year=2024, author=author1)
    book3 = Book(title="AI 윤리의 이해", published_year=2022, author=author2)
    book4 = Book(title="머신러닝 실전", published_year=2023, author=author2)
    
    session.add_all([book1, book2, book3, book4])
    session.commit() # 책 데이터 저장

    print(f"생성된 책: {book1}, {book2}, {book3}, {book4}")

    print("\n--- 데이터 추가 완료! 이제 조회 및 탐색 시작! ---")

    # ----------------------------------------
    # 4. 여기에 데이터 조회 및 관계 탐색 코드를 작성하세요.
    # 예시:
    # found_author = session.get(Author, 1)
    # if found_author:
    #     print(f"작가: {found_author.name}, 쓴 책:")
    #     for book in found_author.books:
    #         print(f"  - {book.title}")
    # ----------------------------------------
    

    # 첫 번째 조회 (완료된 부분)
    found_author = session.get(Author, 1)
    if found_author:
        print(f"\n작가: {found_author.name}, 쓴 책:")
        for book in found_author.books:
            print(f"  - {book.title} ({book.published_year}년 출판)")
    else:
        print("ID 1인 작가를 찾을 수 없습니다.")

    # ----------------------------------------
    # 이제 여기에 나머지 두 가지 조회 코드를 작성해 보세요!

    # 1. 특정 책(ID로 조회)을 찾아서 그 책을 쓴 작가의 이름을 출력해 보세요.
    # 예시: ID가 3인 책 (AI 윤리의 이해)을 찾아 이영희 작가 이름을 출력
    # hint: session.get(Book, 3) 사용 후, book_object.author.name 에 접근!
    print("\n--- 특정 책의 작가 이름 조회 ---")
    # 여기에 코드를 작성하세요.
    found_book = session.get(Book, 1)
    if found_book:
        print(f"작가 이름: {found_book.author.name}")

    

    # 2. 모든 책의 제목과 해당 책의 작가 이름을 함께 출력해 보세요.
    # hint: session.query(Book).all() 사용 후, 각 book_object.author.name 에 접근!
    print("\n--- 모든 책과 작가 조회 ---")
    # 여기에 코드를 작성하세요.
    all_books = session.query(Book).all()
    for book in all_books:
        print(f"책: {book.title}, 작가: {book.author.name}")

    # ----------------------------------------
except Exception as e:
    session.rollback()
    print(f"활동 중 오류 발생: {e}")
finally:
    session.close()
    print("\nAuthor-Book 활동 완료!")
````
